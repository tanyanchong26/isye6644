## Benchmarking

```python
import numpy as np
import matplotlib.pyplot as plt

a = %timeit -o prn.generator_lecuyer([10072020,11072020,12072020], [10072021,11072021,12072021], 100000)
b = %timeit -o prn.generator_desert_island(12072021, 100000)
c = %timeit -o prn.generator_tausworthe(100000)
d = %timeit -o prn.generator_randu(12072021, 100000)
e = %timeit -o [uniform() for i in range(0,100000)]
f = %timeit -o [rng.random() for i in range(0,100000)]

x_pos = np.arange(6)
method = ["L'Ecuyer", "Desert Island", "Tausworthe", "RANDU", "PCG64", "LFSR113"]
mean_time = [a.average, b.average, c.average, d.average, e.average, f.average]
deviation = [a.stdev, b.stdev, c.stdev, d.stdev, e.stdev, f.stdev]

# Build the plot
fig, ax = plt.subplots()

ax.bar(x_pos, mean_time, yerr=deviation, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Computation Time')
ax.set_xticks(x_pos)
ax.set_xticklabels(method)
ax.set_title('Comparison of Run Time Efficiency of Generators')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.show()
```
